```python
rename_dict = ['№ Summer', 'GS', 'SS', 'BS', 'Total Summer', '№ Winter', 'GW', 'SW', 'BW', 'Total Winter', '№ Games', 'GT', 'ST', 'BT', 'Combined total']
df.columns = rename_dict
```

## Data analysis

```python
df.sort_values('GW', ascending = False).iloc[1:6] #sort values bsed on number of Winter Gold
```

```python
df['Weighted TW'] = 3*df['GT'] + 2*df['ST'] + df['BT'] 
df.sort_values('Weighted TW', ascending = False).iloc[1:6,:] #sort values based on number of scores
```
